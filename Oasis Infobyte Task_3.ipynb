# Preparing a model


```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('price~wheelbase+carlength+carwidth+carheight+curbweight+boreratio+stroke+horsepower+citympg+highwaympg',data=df).fit()
```

```python
model.summary()
```

 carwidth, curbweight, stroke, horsepower these variables are significant variable because p value of these variables are lessthan 0.05 

We can drop wheelbase, carLength, boreratio, citympg, highwaympg columns because p value of these columns are greaterthan 0.05

# Calculating VIF

```python
rsq_wheelbas = smf.ols('wheelbase~carlength+boreratio+citympg+highwaympg',data=df).fit().rsquared  
vif_wheelbas = 1/(1-rsq_wheelbas) 

rsq_carlength = smf.ols('carlength~wheelbase+boreratio+citympg+highwaympg',data=df).fit().rsquared  
vif_carlength = 1/(1-rsq_carlength) 

rsq_boreratio = smf.ols('boreratio~wheelbase+carlength+citympg+highwaympg',data=df).fit().rsquared  
vif_boreratio = 1/(1-rsq_boreratio)

rsq_citympg = smf.ols('citympg~wheelbase+carlength+boreratio+highwaympg',data=df).fit().rsquared  
vif_citympg = 1/(1-rsq_citympg) 

rsq_highwaympg = smf.ols('highwaympg~wheelbase+carlength+boreratio+citympg',data=df).fit().rsquared  
vif_highwaympg = 1/(1-rsq_highwaympg)

# Storing vif values in a data frame
d1 = {'Variables':['wheelbase','carlength','boreratio','citympg','highwaympg'],'VIF':[vif_wheelbas,vif_carlength,vif_boreratio,vif_citympg,vif_highwaympg]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

# Residual Analysis

## Test for Normality of Residuals (Q-Q Plot)

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
import numpy as np
list(np.where(model.resid>10))
```

## Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```
